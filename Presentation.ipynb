### The Dynamics of Satisfaction with Working Hours in Australia: The Usefulness of Panel Data in Evaluating the Case for Policy Intervention, by Breunig, Gong and Leslie.    

### Introduction to the paper 
    - Effective public policy interventions are important for a government, need to find the correct metrics
    - This paper suggests Severity, Prevalence, and Persistence of a problem are good metrics to find if an intervention is needed
    - They look at these three metrics with regards to Working Hour diassatifaction in Australia using the HILDA data from 2001-2011. 

### Summary of Results


### Conclusion
    - Desire to work less is much more prevalent than a desire to work more. 
    - A desire to work more mismatch resolved within one year, but unknown how long a desire to work less gets resolved.
    - The study concludes that if people are able to freely change jobs, an hour mismatch does not qualify for any sort of  policy intervention
